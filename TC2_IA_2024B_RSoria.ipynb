```python
# Richard Soria
# Regresión Lineal Simple: Tamaño y precio de la vivienda

# Importar librerías a utilizar
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import matplotlib.pyplot as plt

# 1.- Importar datos
dataframe = pd.read_excel('WSHouse.xlsx')
print(dataframe.head())
print(dataframe.describe())
```

```python
# Separar las caracterísiticas X y las etiquetas Y
x = dataframe[['Tamanio']]
y = dataframe[['Precio_venta']]
```

```python
# 2.- Dividir los conjuntos de datos (70% entrenamiento, 30% prueba)
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
# 3.- Construir el modelo
model = LinearRegression()
```

```python
# 4.- Entrenar el modelo
model.fit(x_train, y_train)

# Evaluar el modelo
y_pred = model.predict(x_test)
```

```python
# 6.- Métricas de evaluación
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print('Error Cuadrático Medio:', mse)
print('Coeficiente de determinación:', r2)
```

```python
# Gráficos de disperción
plt.scatter(x,y)
plt.plot(x_test, y_pred, color='red')
plt.title('Regresión Lineal')
plt.xlabel('Tamaño')
plt.ylabel("Precio de Venta")
```

```python
# Ecuación de regresión estimada para 1500 pies cuadrados
print('Coeficientes:', model.coef_)
print('Intercepción:', model.intercept_)
erm_1500 = model.intercept_ + model.coef_*1.5
print('Ecuación de regresión estimada para 1500 pies cuadrados:', erm_1500)
```

```python
# Ecuación de regresión estimada para 2500 pies cuadrados
erm_2500 = model.intercept_ + model.coef_*2.5
print('Ecuación de regresión estimada para 2500 pies cuadrados:', erm_2500)
```
